# Predictive Maintenance for Fleet Vehicles
## BorgWarner Data Science Project

This project builds a machine learning model to predict vehicle failures using sensor data.

## Problem Statement
Fleet vehicles experience unexpected failures. Goal is to predict failures proactively.

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

## Load Dataset

```
df = pd.read_csv('sample_vehicle_sensor_data.csv')
df.head()
```

## Simple Model

```
X = df[['engine_temp_c','vibration_mm_s','avg_rpm','error_code_count','battery_voltage']]
y = np.random.randint(0,2,len(df))

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train,y_train)

preds = model.predict(X_test)
print(classification_report(y_test,preds))
```
